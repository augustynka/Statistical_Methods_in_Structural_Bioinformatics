
Ocena 3.0

Przeanalizujmy tabelę z kolumnami: PDB id, Chain, GNRA i Description, aby określić następujące
prawdopodobieństwa:

• P (GNRA): Prawdopodobieństwo, że łańcuch zawiera motyw GNRA.

• P (Ribosomal): Prawdopodobieństwo, że łańcuch zawiera rybosomalne RNA.

• P (Riboswitch): Prawdopodobieństwo, że łańcuch zawiera ryboprzełącznik.


```python
import pandas as pd
from fractions import Fraction
import numpy as np

gnra= pd.read_csv("gnra.csv")

count_chains=len(gnra)
count_gnra=gnra['GNRA'].value_counts().get(True,0)
prawdopodobienstwo1 = Fraction(count_gnra,count_chains).limit_denominator()
prawdopodobienstwo1_dziesietnie = count_gnra / count_chains
print("P(GNRA) = ", prawdopodobienstwo1)
print("Reprezentacja dziesiętna ->\tP(GNRA)= ", round(prawdopodobienstwo1_dziesietnie,3))

gnra["Description"]=gnra["Description"].str.lower()
count_ribosomal=gnra["Description"].str.contains("ribosomal").sum()
ribosomal= Fraction(count_ribosomal,count_chains).limit_denominator()
ribosomal_dziesietnie= count_ribosomal / count_chains
print("\nP(Ribosomal) = ", ribosomal)
print("Reprezentacja dziesiętna ->\tP(Ribosomal)= ", round(ribosomal_dziesietnie,3))

count_riboswitch=gnra["Description"].str.contains("riboswitch").sum()
riboswitch=Fraction(count_riboswitch,count_chains).limit_denominator()
riboswitch_dziesietnie=count_riboswitch / count_chains
print("\nP(Riboswitch)= ", riboswitch)
print("Reprezentacja dziesiętna -> \tP(Riboswitch)=", round(riboswitch_dziesietnie,3))
```

```python
import pandas as pd
from fractions import Fraction
import numpy as np

gnra= pd.read_csv("gnra.csv")
gnra["Description"]=gnra["Description"].str.lower()
count_chains=len(gnra)

count_ngnra_czwspolna_nribosomal =len(gnra[(~gnra['Description'].str.contains("ribosomal") & (gnra['GNRA'] == False))])
ngnra_nribosomal = Fraction(count_ngnra_czwspolna_nribosomal,count_chains).limit_denominator()
print("P(¬GNRA∩¬Ribosomal) ", ngnra_nribosomal)
ngnra_nribosomal_dziesietnie= count_ngnra_czwspolna_nribosomal / count_chains
print("Reprezentacja dziesiętna ->\tP(¬GNRA∩¬Ribosomal)= ", round(ngnra_nribosomal_dziesietnie,4))

count_ngnra_czwspolna_ribosomal =len(gnra[(gnra['Description'].str.contains("ribosomal") & (gnra['GNRA'] == False))])
ngnra_ribosomal = Fraction(count_ngnra_czwspolna_ribosomal,count_chains).limit_denominator()
print("P(¬GNRA∩Ribosomal) ", ngnra_ribosomal)
ngnra_ribosomal_dziesietnie= count_ngnra_czwspolna_ribosomal / count_chains
print("Reprezentacja dziesiętna ->\tP(¬GNRA∩Ribosomal)= ", round(ngnra_ribosomal_dziesietnie,4))

count_gnra_czwspolna_nribosomal =len(gnra[(~gnra['Description'].str.contains("ribosomal") & (gnra['GNRA'] == True))])
gnra_nribosomal = Fraction(count_gnra_czwspolna_nribosomal,count_chains).limit_denominator()
print("P(GNRA∩¬Ribosomal) ", gnra_nribosomal)
gnra_nribosomal_dziesietnie= count_gnra_czwspolna_nribosomal / count_chains
print("Reprezentacja dziesiętna ->\tP(GNRA∩¬Ribosomal)= ", round(gnra_nribosomal_dziesietnie,4))

count_gnra_czwspolna_ribosomal =len(gnra[(gnra['Description'].str.contains("ribosomal") & (gnra['GNRA'] == True))])
gnra_ribosomal = Fraction(count_gnra_czwspolna_ribosomal,count_chains).limit_denominator()
print("P(GNRA∩Ribosomal) ", gnra_ribosomal)
gnra_ribosomal_dziesietnie= count_gnra_czwspolna_ribosomal / count_chains
print("Reprezentacja dziesiętna ->\tP(GNRA∩Ribosomal)= ", round(gnra_ribosomal_dziesietnie,4))

#riboswitch

count_ngnra_czwspolna_nriboswitch =len(gnra[(~gnra['Description'].str.contains("riboswitch") & (gnra['GNRA'] == False))])
ngnra_nriboswitch = Fraction(count_ngnra_czwspolna_nriboswitch,count_chains).limit_denominator()
print("\nP(¬GNRA∩¬Riboswitch) ", ngnra_nriboswitch)
ngnra_nriboswitch_dziesietnie= count_ngnra_czwspolna_nriboswitch / count_chains
print("Reprezentacja dziesiętna ->\tP(¬GNRA∩¬Ribowswitch)= ", round(ngnra_nriboswitch_dziesietnie,4))

count_ngnra_czwspolna_riboswitch =len(gnra[(gnra['Description'].str.contains("riboswitch") & (gnra['GNRA'] == False))])
ngnra_riboswitch = Fraction(count_ngnra_czwspolna_riboswitch,count_chains).limit_denominator()
print("P(¬GNRA∩Riboswitch) ", ngnra_riboswitch)
ngnra_riboswitch_dziesietnie= count_ngnra_czwspolna_riboswitch / count_chains
print("Reprezentacja dziesiętna ->\tP(¬GNRA∩Riboswitch)= ", round(ngnra_riboswitch_dziesietnie,4))

count_gnra_czwspolna_nriboswitch =len(gnra[(~gnra['Description'].str.contains("riboswitch") & (gnra['GNRA'] == True))])
gnra_nriboswitch = Fraction(count_gnra_czwspolna_nriboswitch,count_chains).limit_denominator()
print("P(GNRA∩¬Riboswitch) ", gnra_nriboswitch)
gnra_nriboswitch_dziesietnie= count_gnra_czwspolna_nriboswitch / count_chains
print("Reprezentacja dziesiętna ->\tP(GNRA∩¬Riboswitch)= ", round(gnra_nriboswitch_dziesietnie,4))

count_gnra_czwspolna_riboswitch =len(gnra[(gnra['Description'].str.contains("riboswitch") & (gnra['GNRA'] == True))])
gnra_riboswitch = Fraction(count_gnra_czwspolna_riboswitch,count_chains).limit_denominator()
print("(GNRA∩Riboswitch) ", gnra_riboswitch)
gnra_riboswitch_dziesietnie= count_gnra_czwspolna_riboswitch / count_chains
print("Reprezentacja dziesiętna ->\tP(GNRA∩Riboswitch)= ", round(gnra_riboswitch_dziesietnie,3))



gnra["Ribosomal"] = gnra['Description'].str.contains("ribosomal")
gnra["Riboswitch"] = gnra['Description'].str.contains("riboswitch")

crosstab_ribosomal = pd.crosstab(gnra["GNRA"], gnra["Ribosomal"])
macierz_ribosomal = crosstab_ribosomal / count_chains
crosstab_riboswitch = pd.crosstab(gnra["GNRA"], gnra["Riboswitch"])
macierz_riboswitch = crosstab_riboswitch / count_chains
print("\n",macierz_ribosomal)
print("\n",macierz_riboswitch)
```

Ocena 4.0

Wyznacz prawdopodobieństwa warunkowe:

• P (GNRA|Ribosomal)

• P (GNRA|Riboswitch)

• P (Ribosomal|GNRA)

• P (Riboswitch|GNRA)

Wykorzystaj wzory na prawdopodobieństwa warunkowe i twierdzenie Bayesa, aby wykazać, że wartości
te są identyczne.

```python
import pandas as pd
from fractions import Fraction
import numpy as np

gnra= pd.read_csv("gnra.csv")
gnra["Description"]=gnra["Description"].str.lower()
count_chains=len(gnra)
print("Wykniki z prawdopodobieństwa warunkowego")
count_gnra_czwspolna_ribosomal =len(gnra[(gnra['Description'].str.contains("ribosomal") & (gnra['GNRA'] == True))])
count_ribosomal=gnra["Description"].str.contains("ribosomal").sum()
gnra_warun_ribosomal= Fraction(count_gnra_czwspolna_ribosomal,count_ribosomal).limit_denominator()
print("P(GNRA|Ribosomal) = ", gnra_warun_ribosomal)

count_gnra_czwspolna_riboswitch =len(gnra[(gnra['Description'].str.contains("riboswitch") & (gnra['GNRA'] == True))])
count_riboswitch=gnra["Description"].str.contains("riboswitch").sum()
gnra_warun_riboswitch= Fraction(count_gnra_czwspolna_riboswitch,count_riboswitch).limit_denominator()
print("P(GNRA|Riboswitch) = ", gnra_warun_riboswitch)

count_ribosomal_czwspolna_gnra =len(gnra[(gnra['Description'].str.contains("ribosomal") & (gnra['GNRA'] == True))])
count_gnra=gnra['GNRA'].value_counts().get(True,0)
ribosomal_warun_gnra= Fraction(count_ribosomal_czwspolna_gnra,count_gnra).limit_denominator()
print("P(Ribosomal|GNRA) = ", ribosomal_warun_gnra)

count_riboswitch_czwspolna_gnra =len(gnra[(gnra['Description'].str.contains("riboswitch") & (gnra['GNRA'] == True))])
count_gnra=gnra['GNRA'].value_counts().get(True,0)
riboswitch_warun_gnra= Fraction(count_riboswitch_czwspolna_gnra,count_gnra).limit_denominator()
print("P(Riboswitch|GNRA) = ", riboswitch_warun_gnra)

print("\nWykniki z twierdzenia Bayesa")
bayes_gnra_ribosomal = (ribosomal_warun_gnra * count_gnra)/count_ribosomal
print ("P(GNRA|Ribosomal)= ", bayes_gnra_ribosomal)
bayes_gnra_riboswitch = (riboswitch_warun_gnra * count_gnra)/count_riboswitch
print ("P(GNRA|Riboswitch)= ", bayes_gnra_riboswitch)
bayes_ribosomal_gnra = (gnra_warun_ribosomal * count_ribosomal)/count_gnra
print ("P(Ribosomal|GNRA)= ", bayes_ribosomal_gnra)
bayes_riboswitch_gnra = (gnra_warun_riboswitch * count_riboswitch)/count_gnra
print ("P(Riboswitch|GNRA)= ", bayes_riboswitch_gnra)
print("\nUdowodniono, że wartości są identyczne.")
```

Ocena 5

Przygotuj wizualizacje dla czterech rozkładów Beta:

• P (Ribosomal): Beta(liczba rybosomów + 1, liczba struktur - liczba rybosomów + 1)

• P (Riboswitch): Beta(liczba ryboprzełączników + 1, liczba struktur - liczba ryboprzełączników +
1)

• P (GN RA|Ribosomal): Beta(liczba rybosomów z GNRA + 1, liczba rybosomów bez GNRA + 1)

• P (GN RA|Riboswitch): Beta(liczba ryboprzełączników z GNRA + 1, liczba ryboprzełączników
bez GNRA + 1)

Przeprowadź symulację Monte Carlo: dla każdego z rozkładów wykonaj 10 000 losowań wartości.
Oblicz współczynnik Bayesa dla każdej z czwórek wartości i zapisz na liście. Wyznacz średnią, medianę
oraz 95% przedział ufności (określ wartości 2.5 percentyla oraz 97.5 percentyla). Oceń, czy wyniki
symulacji Monte Carlo są zgodne z wcześniej wyznaczoną wartością współczynnika Bayesa.


**Bayes Factor wcześniej wynosił 0.4643, wartość ta niemal idealnie wpisuje się w średnią oraz medianę z symulacji, więc wyniki symulacji Monte Carlo są zgodne z wcześniej wyzwaczoną wartością współczynnika Bayesa oraz wpisują się w przedział ufności.**

```python
import pandas as pd
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt

gnra= pd.read_csv("gnra.csv")
gnra["Description"]=gnra["Description"].str.lower()
count_chains=len(gnra)

count_ribosomal=gnra["Description"].str.contains("ribosomal").sum()
count_gnra_ribosomal =len(gnra[(gnra['GNRA'] == True) & (gnra['Description'].str.contains("ribosomal"))])
count_ngnra_ribosomal = count_ribosomal - count_gnra_ribosomal

count_riboswitch =gnra["Description"].str.contains("riboswitch").sum()
count_gnra_riboswitch = len(gnra[(gnra['GNRA'] == True) & (gnra['Description'].str.contains("riboswitch"))])
count_gnra_riboswitch = count_riboswitch - count_gnra_riboswitch

beta_ribosomal= beta(count_ribosomal +1, count_chains - count_gnra_ribosomal +1)
beta_riboswitch= beta(count_riboswitch + 1, count_chains - count_riboswitch + 1)
beta_gnra_ribosomal= beta(count_gnra_ribosomal + 1, count_ngnra_ribosomal + 1)
beta_gnra_riboswitch= beta(count_gnra_riboswitch + 1, count_gnra_riboswitch + 1)

x = np.linspace(0, 1, 1000)

plt.title("Rozkłady Beta")
plt.plot(x, beta_ribosomal.pdf(x), label="P(Ribosomal)")
plt.plot(x, beta_riboswitch.pdf(x), label="P(Riboswitch)")
plt.plot(x, beta_gnra_ribosomal.pdf(x), label="P(GNRA|Ribosomal)")
plt.plot(x, beta_gnra_riboswitch.pdf(x), label="P(GNRA|Riboswitch)")
plt.legend()

losowania = 10000

ribosomal_monte = beta_ribosomal.rvs(losowania)
riboswitch_monte = beta_riboswitch.rvs(losowania)
gnra_ribosomal_monte = beta_gnra_ribosomal.rvs(losowania)
gnra_riboswitch_monte = beta_gnra_riboswitch.rvs(losowania)

bayes_factors = (gnra_ribosomal_monte  * ribosomal_monte) / (gnra_riboswitch_monte  * riboswitch_monte ) 

bayes_mean = np.mean(bayes_factors)
bayes_median = np.median(bayes_factors)
bayes_lower = np.percentile(bayes_factors, 2.5)
bayes_upper = np.percentile(bayes_factors, 97.5)

print("Średni współczynnik Bayesa:", round(bayes_mean, 4))
print("Mediana współczynnika Bayesa:", round(bayes_median, 4))
print("95% przedział ufności dla współczynnika Bayesa:", (round(bayes_lower, 4), round(bayes_upper, 4)))

plt.figure(figsize=(8, 6))
plt.hist(bayes_factors, bins=50, density=False, alpha=0.7, color='blue')
plt.title("Rozkład symulowanych współczynników Bayesa")
plt.xlabel("Współczynnik Bayesa")
plt.ylabel("Liczba symulacji")
plt.axvline(bayes_mean, color='red', linestyle='dashed', linewidth=1, label=f'Średnia: {round(bayes_mean, 4)}')
plt.axvline(bayes_median, color='orange', linestyle='dashed', linewidth=1, label=f'Mediana: {round(bayes_median, 4)}')
plt.axvline(bayes_lower, color='purple', linestyle='dotted', linewidth=1, label=f'2.5 percentyl: {round(bayes_lower, 4)}')
plt.axvline(bayes_upper, color='purple', linestyle='dotted', linewidth=1, label=f'97.5 percentyl: {round(bayes_upper, 4)}')
plt.legend()
plt.show()
```
